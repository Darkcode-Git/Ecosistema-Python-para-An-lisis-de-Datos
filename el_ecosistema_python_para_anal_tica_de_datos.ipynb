 Modelo de un carrito de compras de una tienda online.

```python
class CarritoCompras:
    def __init__(self, cliente):
        """
        Inicializa un carrito de compras para un cliente.

        Args:
            cliente (str): Nombre del dueño del carrito
        """
        self.cliente = cliente
        self.productos = {}  # diccionario {nombre_producto: precio}

    def agregar(self, producto, precio):
        """
        Añade un producto al carrito.

        Args:
            producto (str): Nombre del producto
            precio (float): Precio del producto
        """
        if precio < 0:
            raise ValueError("El precio no puede ser negativo")
        self.productos[producto] = precio

    def total(self):
        """
        Devuelve la suma de los precios de todos los productos.

        Returns:
            float: Suma total de los precios
        """
        return sum(self.productos.values())

    def __str__(self):
        """
        Devuelve una representación legible del carrito.

        Returns:
            str: Representación en string del carrito
        """
        if not self.productos:
            productos_str = "No hay productos"
        else:
            productos_str = "\n".join([f"  - {prod}: ${precio:.2f}"
                                     for prod, precio in self.productos.items()])

        return (f"Carrito de compras de {self.cliente}\n"
                f"Productos:\n{productos_str}\n"
                f"Total: ${self.total():.2f}")

    def __len__(self):
        """
        Devuelve la cantidad de productos en el carrito.

        Returns:
            int: Número de productos en el carrito
        """
        return len(self.productos)


# Crear instancias de CarritoCompras
carrito1 = CarritoCompras("Felipe Parra")
carrito2 = CarritoCompras("Karin Camacho")

# Agregar productos al primer carrito
carrito1.agregar("Laptop", 1200.50)
carrito1.agregar("Mouse inalámbrico", 25.99)
carrito1.agregar("Teclado mecánico", 89.99)

# Agregar productos al segundo carrito
carrito2.agregar("Smartphone", 699.99)
carrito2.agregar("Cargador", 19.99)
carrito2.agregar("Funda protectora", 15.50)
carrito2.agregar("Audífonos Bluetooth", 79.99)

# Imprimir los carritos usando print() (método __str__)
print("=== Carrito 1 ===")
print(carrito1)
print(f"Cantidad de productos: {len(carrito1)}")
print()

print("=== Carrito 2 ===")
print(carrito2)
print(f"Cantidad de productos: {len(carrito2)}")
print()
```

Escritura de Datos de un Archivo JSON en Python desde una API

```python
import json
import requests
```

```python
# Realizar una solicitud HTTP a una fuente de datos JSON
url = " https://jsonplaceholder.typicode.com/todos"
respuesta_api = requests.get(url)

# Convertir la respuesta en formato JSON a una estructura de datos de Python
lista_tareas = json.loads(respuesta_api.text)
```

Estructura de Basa de Archivos JSON

```python
 # Mostrando los primeros dos elementos de la lista de tareas
print("Primeros dos elementos de la lista de tareas:", lista_tareas[:4])
```

Lectura de Datos de un Archivo JSON.

```python
# Inicializando el diccionario para contabilizar las tareas completadas por usuario
tareas_completadas_por_usuario = {}
```

 Calcular el número de tareas completadas por cada usuario

```python
for tarea in lista_tareas:
    if tarea['completed']:
        user_id = tarea['userId']
        tareas_completadas_por_usuario[user_id] = tareas_completadas_por_usuario.get(user_id, 0) + 1

# Encontrar el número máximo de tareas completadas
max_tareas_completadas = 0
if tareas_completadas_por_usuario:
    max_tareas_completadas = max(tareas_completadas_por_usuario.values())

# Encontrar todos los usuarios que han completado el número máximo de tareas
usuarios_con_mas_tareas = []
for user_id, count in tareas_completadas_por_usuario.items():
    if count == max_tareas_completadas:
        usuarios_con_mas_tareas.append(user_id)

print("Usuarios que han completado la mayor cantidad de tareas TODO:")
for user_id in usuarios_con_mas_tareas:
    print(f"  - Usuario ID: {user_id} con {max_tareas_completadas} tareas completadas")
```

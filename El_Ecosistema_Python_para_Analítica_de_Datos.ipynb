 Modelo de un carrito de compras de una tienda online.

```python
class CarritoCompras:
    def __init__(self, cliente):
        """
        Inicializa un carrito de compras para un cliente.

        Args:
            cliente (str): Nombre del dueño del carrito
        """
        self.cliente = cliente
        self.productos = {}  # diccionario {nombre_producto: precio}

    def agregar(self, producto, precio):
        """
        Añade un producto al carrito.

        Args:
            producto (str): Nombre del producto
            precio (float): Precio del producto
        """
        if precio < 0:
            raise ValueError("El precio no puede ser negativo")
        self.productos[producto] = precio

    def total(self):
        """
        Devuelve la suma de los precios de todos los productos.

        Returns:
            float: Suma total de los precios
        """
        return sum(self.productos.values())

    def __str__(self):
        """
        Devuelve una representación legible del carrito.

        Returns:
            str: Representación en string del carrito
        """
        if not self.productos:
            productos_str = "No hay productos"
        else:
            productos_str = "\n".join([f"  - {prod}: ${precio:.2f}"
                                     for prod, precio in self.productos.items()])

        return (f"Carrito de compras de {self.cliente}\n"
                f"Productos:\n{productos_str}\n"
                f"Total: ${self.total():.2f}")

    def __len__(self):
        """
        Devuelve la cantidad de productos en el carrito.

        Returns:
            int: Número de productos en el carrito
        """
        return len(self.productos)


# Crear instancias de CarritoCompras
carrito1 = CarritoCompras("Felipe Parra")
carrito2 = CarritoCompras("Karin Camacho")

# Agregar productos al primer carrito
carrito1.agregar("Laptop", 1200.50)
carrito1.agregar("Mouse inalámbrico", 25.99)
carrito1.agregar("Teclado mecánico", 89.99)

# Agregar productos al segundo carrito
carrito2.agregar("Smartphone", 699.99)
carrito2.agregar("Cargador", 19.99)
carrito2.agregar("Funda protectora", 15.50)
carrito2.agregar("Audífonos Bluetooth", 79.99)

# Imprimir los carritos usando print() (método __str__)
print("=== Carrito 1 ===")
print(carrito1)
print(f"Cantidad de productos: {len(carrito1)}")
print()

print("=== Carrito 2 ===")
print(carrito2)
print(f"Cantidad de productos: {len(carrito2)}")
print()
```

Escritura de Datos en un Archivo CSV.

```python
# PARTE 2: Escritura de datos en CSV
import csv

frutas = [
    ["Manzana", "Roja",     "Dulce"],
    ["Plátano", "Amarillo", "Dulce"],
    ["Lima",    "Verde",    "Ácida"]
]

ARCHIVO_CSV = "frutas.csv"

def escribir_csv(nombre_archivo, datos):
    """
    Guarda una lista de listas en un archivo CSV.

    Args:
        nombre_archivo (str): Ruta del archivo de salida.
        datos (list[list]): Filas a escribir.
    """
    with open(nombre_archivo, mode="w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerows(datos)
    print(f"Archivo '{nombre_archivo}' guardado correctamente.")

escribir_csv(ARCHIVO_CSV, frutas)
```

Lectura de Datos de un Archivo CSV.

```python
# PARTE 3: Lectura de datos desde CSV
def leer_csv(nombre_archivo):
    """
    Lee un archivo CSV y devuelve su contenido como lista de listas.

    Args:
        nombre_archivo (str): Ruta del archivo a leer.

    Returns:
        list[list[str]]: Filas del archivo.
    """
    datos = []
    with open(nombre_archivo, mode="r", encoding="utf-8") as f:
        reader = csv.reader(f)
        for fila in reader:
            datos.append(fila)
    return datos

datos_frutas = leer_csv(ARCHIVO_CSV)
print(datos_frutas)
```

 Inventar tu propia pipeline de acciones en python sobre el conjunto de datos de tu preferencia en la plataforma.

```python
# PARTE 4: Pipeline con JSONPlaceholder — /users y /todos

# Pipeline:
#   1. GET /users  → obtener todos los usuarios
#   2. GET /todos  → obtener todas las tareas
#   3. Cruzar datos: agregar nombre del usuario a cada tarea
#   4. Filtrar: solo tareas completadas (completed = True)
#   5. POST /todos → simular creación de una nueva tarea
#   6. Mostrar resumen final

import json
import requests

# Definir las URLs base para los endpoints
USERS_API_URL = "https://jsonplaceholder.typicode.com/users"
TODOS_API_URL = "https://jsonplaceholder.typicode.com/todos"

def obtener_usuarios():
    """
    Obtiene la lista de usuarios desde /users.

    Returns:
        list[dict]: Lista de usuarios.
    """
    respuesta = requests.get(USERS_API_URL)
    respuesta.raise_for_status() # Lanza un error para códigos de estado HTTP erróneos
    return respuesta.json()

def obtener_tareas():
    """
    Obtiene todas las tareas desde /todos.

    Returns:
        list[dict]: Lista de tareas.
    """
    respuesta = requests.get(TODOS_API_URL)
    respuesta.raise_for_status() # Lanza un error para códigos de estado HTTP erróneos
    return respuesta.json()

def cruzar_usuario_tarea(usuarios, tareas):
    """
    Agrega el nombre del usuario a cada tarea usando userId como llave.

    Args:
        usuarios (list[dict]): Lista de usuarios.
        tareas (list[dict]): Lista de tareas.

    Returns:
        list[dict]: Tareas enriquecidas con el nombre del usuario.
    """
    # Construye un mapa {userId: nombre} para búsqueda eficiente
    mapa_usuarios = {u["id"]: u["name"] for u in usuarios}

    for tarea in tareas:
        tarea["nombre_usuario"] = mapa_usuarios.get(tarea["userId"], "Desconocido")

    return tareas

def filtrar_completadas(tareas):
    """
    Filtra y devuelve solo las tareas con completed = True.

    Args:
        tareas (list[dict]): Lista de tareas.

    Returns:
        list[dict]: Tareas completadas.
    """
    return [t for t in tareas if t["completed"]]

def crear_tarea(user_id, titulo):
    """
    Simula la creación de una nueva tarea mediante POST /todos.

    Args:
        user_id (int): ID del usuario asignado.
        titulo (str): Título de la nueva tarea.

    Returns:
        dict: Respuesta de la API con la tarea creada.
    """
    nueva_tarea = {
        "userId": user_id,
        "title": titulo,
        "completed": False
    }
    respuesta = requests.post(TODOS_API_URL, json=nueva_tarea)
    respuesta.raise_for_status() # Lanza un error para códigos de estado HTTP erróneos
    return respuesta.json()

def mostrar_resumen(tareas_completadas, nueva_tarea):
    """
    Imprime un resumen del pipeline: conteo por usuario y nueva tarea.

    Args:
        tareas_completadas (list[dict]): Tareas filtradas.
        nueva_tarea (dict): Tarea creada vía POST.
    """
    print(f"Total de tareas completadas: {len(tareas_completadas)}\n")

    # Conteo de tareas completadas por usuario
    conteo = {}
    for t in tareas_completadas:
        nombre = t["nombre_usuario"]
        conteo[nombre] = conteo.get(nombre, 0) + 1

    print("Tareas completadas por usuario:")
    for nombre, cantidad in sorted(conteo.items(), key=lambda x: -x[1]):
        print(f"  {nombre}: {cantidad}")

    print(f"\nNueva tarea creada:")
    print(f"  ID     : {nueva_tarea['id']}")
    print(f"  userId : {nueva_tarea['userId']}")
    print(f"  Título : {nueva_tarea['title']}")
    print(f"  Estado : {'Completada' if nueva_tarea['completed'] else 'Pendiente'}")

def ejecutar_pipeline():
    """
    Orquesta y ejecuta todos los pasos del pipeline de datos.
    """
    print("--- Ejecutando Pipeline de JSONPlaceholder ---")

    try:
        # 1. Obtener usuarios
        usuarios = obtener_usuarios()
        print(f"Paso 1: Se obtuvieron {len(usuarios)} usuarios.")

        # 2. Obtener tareas
        tareas = obtener_tareas()
        print(f"Paso 2: Se obtuvieron {len(tareas)} tareas.")

        # 3. Cruzar datos: agregar nombre del usuario a cada tarea
        tareas_con_nombres = cruzar_usuario_tarea(usuarios, tareas)
        print("Paso 3: Nombres de usuario agregados a las tareas.")

        # 4. Filtrar: solo tareas completadas
        tareas_completadas = filtrar_completadas(tareas_con_nombres)
        print(f"Paso 4: Se filtraron {len(tareas_completadas)} tareas completadas.")

        # 5. POST /todos → simular creación de una nueva tarea
        # Usaremos un userId que sabemos que existe, por ejemplo, el primero de la lista de usuarios.
        simulated_user_id = 1 # O se podría obtener de la lista de usuarios
        nueva_tarea = crear_tarea(simulated_user_id, "Estudiar para el examen de Python avanzado")
        print(f"Paso 5: Se simuló la creación de una nueva tarea con ID {nueva_tarea['id']}.")

        # 6. Mostrar resumen final
        mostrar_resumen(tareas_completadas, nueva_tarea)
        print("\n--- Pipeline Finalizado ---")

    except requests.exceptions.RequestException as e:
        print(f"Error al conectar con la API: {e}")
    except Exception as e:
        print(f"Ocurrió un error inesperado durante el pipeline: {e}")

# Ejecutar el pipeline
ejecutar_pipeline()
```

Aquí tienes un resumen de lo que ha sucedido:
Obtención de datos: Se obtuvieron 10 usuarios y 200 tareas de la API de JSONPlaceholder.
Cruce de datos: Se agregó el nombre de cada usuario a sus tareas correspondientes.
Filtrado: Se identificaron 90 tareas completadas.
Resumen final: Se muestra el total de tareas completadas y un desglose por usuario, destacando la cantidad de tareas completadas por cada uno.
Este resultado demuestra que el pipeline está funcionando como se esperaba, procesando los datos de la API y generando un informe útil.
